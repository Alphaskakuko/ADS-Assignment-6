```python
import plotly.express as px

# Load Gapminder data into a Pandas DataFrame
df = px.data.gapminder()

# Filter data to include only Asian countries
df_asia = df[df['continent'] == 'Asia']

# Calculate the mean life expectancy of Asian countries
mean_life_exp = df_asia['lifeExp'].mean()

print('The mean life expectancy of all the Asian countries is:', round(mean_life_exp, 2))
```

```python
import plotly.express as px

# Load Gapminder data into a Pandas DataFrame
df = px.data.gapminder()

# Filter data to include only European and South American countries
df_europe_south_america = df[df['continent'].isin(['Europe', 'South America'])]

# Calculate the standard deviation of GDP per capita for each country
std_gdp = df_europe_south_america.groupby('country')['gdpPercap'].std()

print('Standard deviation of GDP per capita for each country in Europe and South America:')
print(std_gdp)
```

```python
import plotly.express as px

# Load Gapminder data into a Pandas DataFrame
df = px.data.gapminder()

# Filter data to include only African countries and the years between 1970 and 2000
df_africa = df[(df['continent'] == 'Africa') & (df['year'].isin([1970, 1980, 2000]))]

# Calculate the percentage change in population for each country over the three decades
df_africa_pivot = df_africa.pivot(index='country', columns='year', values='pop')
df_africa_pct_change = df_africa_pivot.pct_change(axis='columns').iloc[:, 1:]

print('Percentage change in population for each African country over the last 3 decades:')
print(df_africa_pct_change)
```
